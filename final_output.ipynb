```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('train.csv')
df1.head()
```

```python
from sklearn.model_selection import train_test_split

# Define features and target variables
X = df1.drop(columns=['respondent_id', 'xyz_vaccine', 'seasonal_vaccine'])
y = df1[['xyz_vaccine', 'seasonal_vaccine']]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=707)
```

```python
X.head()
```

```python
df2 = pd.read_csv('test.csv')
df2.head()
```

```python
# Keeping only the columns of train input in test
df2_filtered = df2[X.columns]

df2_filtered.head()
```

## Best Model traing and prediction as per train.ipynb

```python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import roc_auc_score

# Initialize the model
gb_model = MultiOutputClassifier(HistGradientBoostingClassifier(random_state=42))

# Train the model
gb_model.fit(X_train, y_train)

# Make predictions
gb_pred = gb_model.predict_proba(X_test)

# Extract the probabilities
xyz_vaccine_pred = gb_pred[0][:, 1]
seasonal_vaccine_pred = gb_pred[1][:, 1]

# Calculate the ROC AUC score
xyz_roc_auc = roc_auc_score(y_test['xyz_vaccine'], xyz_vaccine_pred, average='macro')
seasonal_roc_auc = roc_auc_score(y_test['seasonal_vaccine'], seasonal_vaccine_pred, average='macro')
print(xyz_roc_auc, seasonal_roc_auc)

# Compute the mean ROC AUC score
mean_roc_auc = (xyz_roc_auc + seasonal_roc_auc) / 2

print(f'HistGradientBoosting Mean ROC AUC Score: {mean_roc_auc}')
```

```python
# predictions on test data
# Make predictions
final_pred = gb_model.predict_proba(df2_filtered)
```

```python
# Extract the probabilities
h1n1_vaccine = final_pred[0][:, 1]
seasonal_vaccine = final_pred[1][:, 1]
```

```python
print(h1n1_vaccine, seasonal_vaccine)
```

```python
respondent_id = df2['respondent_id']
```

```python
data = {
    'h1n1_vaccine': h1n1_vaccine,
    'seasonal_vaccine': seasonal_vaccine
}

df = pd.DataFrame(data)
df.index = respondent_id
df.head()
```
