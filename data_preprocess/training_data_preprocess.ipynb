```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df1 = pd.read_csv('training_set_features.csv')
df2 = pd.read_csv('training_set_labels.csv')
data = pd.merge(df1, df2, on='respondent_id')

data.head()
```

```python
data.info()
```

```python
data = data.drop(['hhs_geo_region', 'employment_industry', 'employment_occupation'], axis=1)
```

```python
data.isnull().sum()
```

```python
# Fill missing values with the NA for categorical columns
for col in data.select_dtypes(include=['object']).columns:
    data[col].fillna('NA', inplace=True)

# Fill missing values with -1 for numerical columns
for col in data.select_dtypes(include=['number']).columns:
    data[col].fillna(-1, inplace=True)
    
data.isnull().sum()
```

```python
data.head()
```

```python
data.set_index('respondent_id', inplace=True)
```

```python
# Identify numerical columns with more than 2 unique values
numerical_cols = data.select_dtypes(include=['number']).columns
numerical_categorical_cols = [col for col in numerical_cols if data[col].nunique() > 2]

# Convert these columns to categorical
for col in numerical_categorical_cols:
    data[col] = data[col].astype('category')
```

```python
from sklearn.preprocessing import OneHotEncoder

# One-hot encode categorical features
categorical_cols = data.select_dtypes(include=['category', 'object']).columns
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
data.head()
```

```python
data.columns
```

```python
# Calculate the correlation matrix
corr_matrix = data.corr().abs()


# # Plot the correlation matrix
# plt.figure(figsize=(10, 8))
# sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
# plt.title('Correlation Matrix')
# plt.show()
```

```python
# Identify columns with a correlation greater than a specified threshold
threshold = 0.7
upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

# Drop the highly correlated features
data.drop(columns=to_drop, inplace=True)
data.head()
```

```python
data.to_csv('train.csv')
```
